### What is Matplotlib?

Matplotlib is powerful plotting library in Python used for creating static, animated and interactive visualization. Matplotlib's primary purpose is to provide users with the tools and functionality to represents data graphically, making it easiar to analyze and understand.

### What is Matplotlib figure?

In Matplotlib, a figure is the top-level container that holds all the elements of a plot. It represents the entire window or page where the plot is drawn.



##### Basic components or parts of Matplotlib Figure

1. Axes: Axes are the rectangular areas within the figure where data is plotted. Each figure contain one or more axes.

2. Axis: Axis objects represents the x-axis and y-axisof the plot. They define the data limits and axis labels.

3. Markers: Markers are symbol used to denote individual data points on a plot. They can be in shape such as circle, square, triangle or custom symbols.

4. Adding Lines to Figures: Lines connect data point on a plot and are commonly used in line plots, scatter plots with connected points and other types of plots. They represent the relationship or trent between data points and can be styled with different colors, widths and styles to convey additional information.

5. Title: The title is the text element that provides a descriptive title for the plot. 

6. Axis Labels: Labels are text elements that provide descriptions for the x-axis and y-axis. They help identify the data being plotted and provide units or other relevant information.

7. Legend: Legent provide a key to the symbols or color used in the plot to represents different data series or categories



### Advantages og Matplotlib

1. **Versatility:** Matplotlib can create a wide range of plots, including line plots, scatter plots, bar plots, histograms, pie charts, and more.
2. **Customization:** It offers extensive customization options to control every aspect of the plot, such as line styles, colors, markers, labels, and annotations.
3. **Integration with NumPy:** Matplotlib integrates seamlessly with NumPy, making it easy to plot data arrays directly.
4. **Publication Quality:** Matplotlib produces high-quality plots suitable for publication with fine-grained control over the plot aesthetics.
5. **Wide Adoption:** Due to its maturity and flexibility, Matplotlib is widely adopted in the scientific and engineering communities.
6. **Extensible:** Matplotlib is highly extensible, with a large ecosystem of add-on toolkits and extensions like Seaborn, Pandas plotting functions, and Basemap for geographical plotting.
7. **Cross-Platform:** It is platform-independent and can run on various operating systems, including Windows, macOS, and Linux.

# 1. Line Plot

A line plot is a type of graph that shows information using a series of data points (usually connected by straight lines). It is commonly used to display continuous data over a time or to show trends, patterns and replationships between two variables. In a line plot:

    • The x-axis represents the independent variable (e.g. time, categories or sequence of events).
    • The y-axis represents the dependent variable (the value you are measuring).

Line plots are commonly used in **time series analysis**, showing a variable changes over time. (e.g., stock prices, temperature, etc.)


##### Example Use Cases for Line Plots:

1. Visualizing trends over a time (e.g., daily temerature, monthly sales).
2. Showing the relationship between two continuous variables (e.g., distance vs time for a moving objects).
3. Representing mathematical functions (e.g., plotting a sine wave).

### How to plot a Line Plot with Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]
```

3. Create the Line Plot: Use the plt.plot() funcgtion to create the line plot. Pass the x and y data to it.

```python
plt.plot(x,y)
```

5. Display the Plot: Finally, use plt.show() to display the plot.

```python
plt.show()
```

##### Full Example:

```python
import matplotlib.pyplot as plt

# data
x = [1,2,3,4,5]
y = [1,4,9,16,25]

# create the line plot
plt.plot(x,y)

# customize the plot
plt.title("Example of Line plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)

# show the plot
plt.show()
```

# 5. Histograms

A Histogram is a type of plot used to represent the distribution of a dataset. It visualizes the frequency or count of data points that fall within certain ranges (called bins or intervals). Each bar in a histogram represents the frequency of data points in a specific range.

##### Key Characteristics of a Histogram:
    • Bins: The data range is divided into intervals (bins). Each bin represents a range of values.
    • Frequency: The height of each bar represents how many data points fall into the corresponding bin.
    • Continuous Data: Histograms are particularly useful for continuous numerical data, where the data can take any value within a given range.

Histograms are especially helpful for understanding the distribution of a dataset, such as whether it is normal, skewed or has multiple peaks (modes).

##### Example Use Cases for Histograms:

1. Visualize the distribution of dataset (e.g., normal, skewed, bimodal).
2. Identify pattern such as the spread, central tendancy or presence of outliers in the data.
3. Compare the frequency of data values in different range.

### How to plot a Histograms with Matplotlib

1. Import Matplotlib: First you need to import the **matplotlib.pyplot** module (commonly imported as plt).

```python
import matplotlib.pyplot as plt
```

2. Prepare the Data: Define a dataset (list, array or series) that you want to visualize as a Histogram.

```python
age = [23,24,25,26,27,28,31,32,33,34,35,50,56,40,28,36,41,42,43,44,45,22,38,42,46,47,48,52,54,56,58,27,44,60]
```

3. Create the Histogram: Use plt.hist() to create the histogram. The basic syntax is:

```python
plt.hist(age, bins=9)
```

4. Customize the plot: Add labels, title and gridlines to enhance the readability of the histogram.

```python
plt.title("Histogram Example")
plt.xlabel("Age")
plt.ylabel("Counts")
plt.grid(True)
```

5. Display the Plot: Display the histogram using plt.show().

```python
plt.show()
```

##### Full Example:

```python
import matplotlib.pyplot as plt

# data
age = [23,24,25,26,27,28,31,32,33,34,35,50,56,40,28,36,41,42,43,44,45,22,38,42,46,47,48,52,54,56,58,27,44,60]

# plot the histogram
plt.hist(age, bins=9)

# customize the plot
plt.title("Histogram Example")
plt.xlabel("Age")
plt.ylabel("Counts")
plt.grid(True)

# display the plot
plt.show()
```

##### Customizing the Histogram:

1. Specifying Bin Ranges: Instad of specifying the number of bins, you can specify the exact bin edges. For example:

```python
bins = [20,25,30,35,40,45,50,55,60]
```

This will devide the data into intervals: [20-25][25-30][30-35][35-40][40-45][45-50][50-55][55-60]

2. Add a Kernel Density Estimate (EDE): You can overlay a KDE (smooth curve) on top of the histogram to visualize the probability density of the data. In Seaborn, the KDE is included by default, but you can also add it in Matplotlib with scipy or seaborn.

```python
import seaborn as sns
sns.histplot(age, bins=9, kde=True)
```

```python
plt.hist(age, bins=18, color='orange', edgecolor='black')
```

```python
plt.hist(age, bins=18, density=True, edgecolor='black')
```

##### Full Example:

```python
import matplotlib.pyplot as plt
import seaborn as sns

# data
age = [23,24,25,26,27,28,31,32,33,34,35,50,56,40,28,36,41,42,43,44,45,22,38,42,46,47,48,52,54,56,58,27,44,60]

# plot the histogram with customization
plt.hist(age, bins=18, color='green', edgecolor='black')
sns.histplot(age, bins=18, kde=True)

# customize the plot
plt.title("Histogram Plot")
plt.xlabel("Age")
plt.ylabel("Counts")
plt.grid(True)

# display the plot
plt.show()
```

# 6. Box Plot

A box plot is a graphical representation of the distribution of numerical data through their quartiles. It highlights the median, upper and lower quartiles, and any outliers. Box plots are useful for showing the spread and skewness of data, as well as identifying any potential outliers.

##### Key Components of a Box Plot:
1. **Box:** Represents the interquartile range (IQR), which contains the middle 50% of the data (from Q1 to Q3).
2. **Whiskers:** Lines extending from the box, indicating the range of data excluding outliers. The whiskers typically extend to 1.5 times the IQR from the box.
3. **Median Line:** A line within the box that shows the median (50th percentile) of the data.
4. **Outliers:** Data points that fall outside the whiskers (typically 1.5 times the IQR away from the quartiles).

##### Example Use Cases for a Box Plot:
1. To understand the spread and central tendency of a dataset.
2. To compare distributions across different categories.
3. To identify outliers in the data.
4. To compare datasets with skewed or asymmetric distributions.

### How to plot a Box Plot with Matplotlib

1. Import Matplotlib: Import the **matplotlib.pyplot** module.

```python
import matplotlib.pyplot as plt
```

2. Prepare the Data: Define the dataset that you want to visualize. This could be a list of values or multiple categories.

```python
data = [ [28, 32, 29, 23, 20],  # Group 1 (e.g., ages of group 1)
         [25, 31, 24, 29, 33],  # Group 2
         [27, 35, 30, 29, 21] ] # Group 3
```

3. Create a Box Plot: Use the plt.boxplot() function to create the box plot.

```python
plt.boxplot(data)
```

4. Customize the Plot: Add titles, labels, and other customizations to the plot.

```python
plt.title("Box Plot Example")
plt.xlabel("Groups")
plt.ylabel("Values")
plt.grid(True)
```

5. Display the Plot: Use plt.show() to display the box plot.

```python
plt.show()
```

##### Full Example:

```python
import matplotlib.pyplot as plt

# data for 3 groups
data = [ [28, 32, 29, 23, 20],  # Group 1
         [25, 31, 24, 29, 33],  # Group 2
         [27, 35, 30, 29, 21] ] # Group 3

# Create the box plot
plt.boxplot(data)

# Customize the plot
plt.title("Box Plot Example")
plt.xlabel("Groups")
plt.ylabel("Values")

# Display the plot
plt.show()
```

##### Customizing the Box Plot:

1. Changing Colors: You can change the color of the boxes, whiskers, and other components.

```python
plt.boxplot(data, boxprops=dict(color="blue"), whiskerprops=dict(color="green"), capprops=dict(color="red"))
```

2. Displaying Notches: A notch can be added to the box plot to show a confidence interval around the median.

```python
plt.boxplot(data, notch=True)
```

3. Changing Whisker Length: The whiskers can be adjusted by changing the whiskerprops parameter.

```python
plt.boxplot(data, widths=0.5, patch_artist=True)
```

4. Adding Data Points: You can overlay individual data points to make the box plot more informative.

```python
plt.boxplot(data, positions=[1, 2, 3])
for i in range(3):
    for j in range(len(data[i])):
        plt.text(i + 1, data[i][j], str(data[i][j]), horizontalalignment='center')
```

##### Full Example:

```python
import matplotlib.pyplot as plt

# data for 3 groups
data = [ [28, 32, 29, 23, 20],  # Group 1
         [25, 31, 24, 29, 33],  # Group 2
         [27, 35, 30, 29, 21] ] # Group 3

# Create the box plot with customization
plt.boxplot(data, patch_artist=True, boxprops=dict(facecolor='lightblue', color='blue'),
            whiskerprops=dict(color='green'), capprops=dict(color='red'), medianprops=dict(color='orange'))

# Add titles and labels
plt.title("Box Plot with Customizations")
plt.xlabel("Groups")
plt.ylabel("Values")

# Display the plot
plt.show()
```

# 7. Pie Chart

A pie chart is a circular graph divided into slices to illustrate numerical proportions. Each slice of the pie represents a category's contribution to the total, making it easy to compare parts of a whole. Pie charts are best used when you want to visualize the relative sizes of different categories within a single dataset.

##### Key Components of a Pie Chart:
1. **Slices:** Each slice represents a category or a segment of the total data.
2. **Percentage Labels:** These labels indicate the percentage contribution of each slice relative to the whole.
3. **Legend:** An optional legend can be added to label each slice with its corresponding category name.

##### Example Use Cases for a Pie Chart:
1. To show the relative proportions of categories within a whole.
2. For visualizing market share, budget distribution, or other scenarios where parts of a total need to be compared.
3. When you want to emphasize the relationship between different categories and their contribution to a total sum.

### How to plot a Pie Chart with Matplotlib

1. Import Matplotlib: First, you need to import the **matplotlib.pyplot** module.

```python
import matplotlib.pyplot as plt
```

2. Prepare the Data: Define the data for the pie chart, which typically consists of numerical values and category labels.

```python
sizes = [25, 40, 15, 20]  # Values (percentages or counts)
labels = ['A', 'B', 'C', 'D']  # Categories
```

3. Create the Pie Chart: Use the plt.pie() function to generate the pie chart.

```python
plt.pie(sizes, labels=labels)
```

4. Customize the Plot: You can add title, customize the appearance, and display percentages on the slices.

```python
plt.title("Pie Chart Example")
```

5. Display the Plot: Use plt.show() to display the pie chart.

#  8. Heatmap

A heatmap is a data visualization technique that uses colors to represent the values in a matrix or a 2D grid. It is especially useful for visualizing complex data in a way that highlights patterns, correlations, and trends. The heatmap uses color gradients to represent different ranges of values, which makes it easy to spot areas of high and low intensity.

##### Key Components of a Heatmap:
1. **Cells:** The heatmap consists of cells arranged in a grid, where each cell represents a value in the dataset.
2. **Color Scale:** The color of each cell represents the magnitude of the value, with darker or lighter colors indicating higher or lower values, respectively.
3. **Axes:** The x-axis and y-axis represent the rows and columns of the dataset, respectively.
4. **Color Bar:** A color bar is usually added to the side of the heatmap to provide a reference for what each color represents in terms of value.

##### Example Use Cases for a Heatmap:
1. To visualize correlations in a matrix of values (e.g., a correlation matrix).
2. For comparing the intensity of values in a dataset, such as geographical or temporal patterns.
3. In clustering applications where you want to visualize similarities or dissimilarities between items.
4. To identify areas of concentration or outliers within a 2D dataset.

# 9. Violin Plot

A violin plot is a data visualization that combines aspects of a box plot and a kernel density plot. It shows the distribution of a dataset across different categories, with the added benefit of visualizing the probability density of the data at different values. The name "violin plot" comes from the shape, which resembles the contours of a violin.

##### Key Components of a Violin Plot:
1. **Density Plot:** The shaded area represents the distribution of the data, with wider areas showing higher density of values.
2. **Box Plot Elements:** Inside the violin plot, you’ll typically see a box plot that shows the median, interquartile range (IQR), and potential outliers.
3. **Individual Data Points (optional):** Some violin plots display individual data points, especially if the dataset isn’t too large.

##### Example Use Cases for Violin Plots:
1. **Comparing Distributions:** Violin plots are excellent for comparing the distributions of multiple categories or groups.
2. **Visualizing Data Distribution:** Useful when you want to see the overall distribution of data along with summary statistics (e.g., the median, quartiles, etc.).
3. **Univariate Data:** Great for visualizing the distribution of a single variable across different groups or categories.

### How to Create a Violin Plot with Seaborn

1. Import Libraries: First, you need to import the matplotlib and seaborn libraries.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

2. Prepare the Data: For a violin plot, you usually need categorical data and corresponding numerical data.

```python
# Sample data
np.random.seed(0)
category = ['A', 'B', 'C']
data = {
    'A': np.random.normal(0, 1, 100),
    'B': np.random.normal(2, 1.5, 100),
    'C': np.random.normal(-2, 0.5, 100)
}
```

3. Create the Violin Plot: Use sns.violinplot() to create the plot. You can pass in the data and specify categories.

```python
df = pd.DataFrame({
    'Category': np.repeat(category, 100),  # Repeat categories for 100 observations each
    'Value': np.concatenate([data['A'], data['B'], data['C']])  # Concatenate all data
})

# Create the violin plot
plt.figure(figsize=(8, 6))
sns.violinplot(x='Category', y='Value', data=df, palette="Set2")
```

4. Customize the Plot: Add titles, labels, and adjust the appearance as needed.

```python
plt.title("Violin Plot Example")
plt.xlabel("Category")
plt.ylabel("Value")
```

5. Display the Plot: Finally, use plt.show() to display the plot.

```python
plt.show()
```

##### Full Example:

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Sample data (random normal distributions for each category)
np.random.seed(0)
category = ['A', 'B', 'C']
data = {
    'A': np.random.normal(0, 1, 100),
    'B': np.random.normal(2, 1.5, 100),
    'C': np.random.normal(-2, 0.5, 100)
}

# Create the violin plot
sns.violinplot(data=[data['A'], data['B'], data['C']], ax=plt.gca())

# Customize the plot
plt.title("Violin Plot Example")
plt.xlabel("Category")
plt.ylabel("Value")

# Display the plot
plt.show()
```

```python
# Adding a 'Group' column to the DataFrame (you can customize it as needed)
df['Group'] = np.random.choice(['X', 'Y'], size=df.shape[0])

sns.violinplot(x='Category', y='Value', hue='Group', data=df, split=True, palette="Set2")
```

4. Adding Individual Data Points: Use swarmplot or stripplot to overlay individual data points on top of the violin plot.

```python
sns.violinplot(data=[data['A'], data['B'], data['C']])
sns.swarmplot(data=[data['A'], data['B'], data['C']], color='k', alpha=0.6)
```

5. Adjusting the Width of the Violin: The scale parameter allows you to adjust the width of the violins, such as setting them to have equal width (scale='count'), which will make the violins the same width regardless of the number of observations.

```python
sns.violinplot(data=[data['A'], data['B'], data['C']], scale='count')
```

6. Rotating X-axis Labels: When you have long category names, you may need to rotate the labels for better readability.

```python
plt.xticks(rotation=45)
```

##### Full Example:

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Sample data
np.random.seed(0)
category = ['A', 'B', 'C']
data = {
    'A': np.random.normal(0, 1, 100),
    'B': np.random.normal(2, 1.5, 100),
    'C': np.random.normal(-2, 0.5, 100)
}

# Create the violin plot with customizations
plt.figure(figsize=(8, 6))
sns.violinplot(data=[data['A'], data['B'], data['C']], palette="Set2", scale='count')

# Overlay individual data points using stripplot
sns.stripplot(data=[data['A'], data['B'], data['C']], color='black', alpha=0.6, jitter=True)

# Customize the plot
plt.title("Customized Violin Plot")
plt.xlabel("Category")
plt.ylabel("Value")
plt.xticks([0, 1, 2], category)  # Set x-ticks to categories
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```

# 10. 3D Plot

A 3D plot is a three-dimensional graph that allows you to visualize data with three variables. In contrast to 2D plots, which represent data on a flat plane, 3D plots provide an additional axis (usually the z-axis), allowing for a more detailed view of data. 3D plots are ideal for visualizing surfaces, 3D scatter plots, and relationships between three continuous variables.

##### Types of 3D Plots:
1. **3D Scatter Plot:** Displays data points in 3D space to show relationships between three variables.
2. **3D Surface Plot:** Represents data points as a continuous surface, typically used to visualize functions of two variables.
3. **3D Wireframe Plot:** Similar to the surface plot, but displays the surface as a grid of lines rather than a solid surface.

##### Key Components of a 3D Plot:
1. **Axes:** The x, y, and z axes represent three different variables.
2. **Data Points:** Points in 3D space or a surface, depending on the type of plot.
3. **Labels:** Axes labels and a title to describe the plot.

##### Example Use Cases for 3D Plots:
1. To visualize the relationship between three continuous variables (e.g., in machine learning or multivariate analysis).
2. To plot functions that depend on two independent variables (e.g., topography, mathematical surfaces).
3. To explore complex relationships or patterns in multidimensional data.

### How to Create a 3D Plot in Matplotlib

1. Import Libraries: You'll need matplotlib and numpy for creating 3D plots.

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

2. Prepare the Data: You can generate or define your 3D data. For example, let's create a 3D scatter plot or surface plot.

Example 1: 3D Scatter Plot, A 3D scatter plot shows individual data points in 3D space.

3. Create the Plot: Initialize a figure and an axis for the 3D plot, then plot the points.

```python
fig = plt.figure()  # Create a figure
ax = fig.add_subplot(111, projection='3d')  # Add 3D subplot

# Data (example)
x = np.random.rand(100)  # x data
y = np.random.rand(100)  # y data
z = np.random.rand(100)  # z data

# Create the 3D scatter plot
ax.scatter(x, y, z)

# Customize the plot
ax.set_title("3D Scatter Plot Example")
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")

# Display the plot
plt.show()
```

Example 2: 3D Surface Plot, A 3D surface plot represents a continuous surface in 3D space.

3. Create the Plot: In this case, you’ll generate a grid of points and use a function to define the z-values.

```python
# Data for 3D Surface Plot
x = np.linspace(-5, 5, 100)  # X values
y = np.linspace(-5, 5, 100)  # Y values
x, y = np.meshgrid(x, y)  # Create a meshgrid
z = np.sin(np.sqrt(x**2 + y**2))  # Z values (function of x and y)

# Create the figure and 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(x, y, z, cmap='viridis')

# Customize the plot
ax.set_title("3D Surface Plot Example")
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")

# Display the plot
plt.show()
```

Example 3: 3D Wireframe Plot, A wireframe plot is similar to a surface plot, but it represents the surface as a grid of lines rather than a solid surface.

3. Create the Plot: Use plot_wireframe instead of plot_surface.

```python
# Data for 3D Wireframe Plot
x = np.linspace(-5, 5, 100)  # X values
y = np.linspace(-5, 5, 100)  # Y values
x, y = np.meshgrid(x, y)  # Create a meshgrid
z = np.sin(np.sqrt(x**2 + y**2))  # Z values (function of x and y)

# Create the figure and 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the wireframe
ax.plot_wireframe(x, y, z, color='black')

# Customize the plot
ax.set_title("3D Wireframe Plot Example")
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")

# Display the plot
plt.show()
```

##### Customizing the 3D Plot:

1. Changing Colors: You can change the color map for the surface plot using the cmap parameter.

```python
ax.plot_surface(x, y, z, cmap='plasma')  # Change the color map
```

2. Adding Titles and Labels: You can customize the axis labels and add a title to the plot.

```python
ax.set_title("My Custom 3D Plot")
ax.set_xlabel("X Axis Label")
ax.set_ylabel("Y Axis Label")
ax.set_zlabel("Z Axis Label")
```

3. Adjusting View Angle: You can rotate the 3D plot to get a better view of the data using the view_init method.

```python
ax.view_init(azim=45, elev=30)  # Set the azimuth and elevation angles
```

4. Adding Gridlines: By default, the grid is shown, but you can adjust it or turn it off.

```python
ax.grid(True)  # Turn on grid
ax.grid(False)  # Turn off grid
```

##### Full Example:

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Data for 3D Surface Plot
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create the figure and 3D axis
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface with color map 'viridis'
surf = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')

# Customize the plot
ax.set_title("3D Surface Plot Example")
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")

# Adding a color bar to indicate the scale
fig.colorbar(surf)

# Adjust the view angle
ax.view_init(azim=60, elev=30)

# Display the plot
plt.show()
```
